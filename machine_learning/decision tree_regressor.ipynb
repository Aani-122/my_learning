# Decision Tree 

```python
import pandas as pd
```

```python
m={
    "ferature":[1,2,3,4,5],
    "target":[2,3,4,5,6]
}
```

```python
df=pd.DataFrame(m)
```

```python
df
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model=DecisionTreeRegressor()
```

```python
X=df[["ferature"]]
y=df["target"]
```

```python
model.fit(X,y)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(model,filled=True)
```

# how pre-prunning hyperparameter going to affect the DT 

# min_samples_leaf

```python
min_samples_leaf_model=DecisionTreeRegressor(min_samples_leaf=2)
```

```python
min_samples_leaf_model.fit(X,y)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(10,5))
tree.plot_tree(min_samples_leaf_model,filled=True)
```

# max_depth

```python
max_depth_model=DecisionTreeRegressor(max_depth=2)
```

```python
max_depth_model.fit(X,y)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(10,5))
tree.plot_tree(max_depth_model,filled=True)
```

# min_samples_split

```python
min_samples_split_model=DecisionTreeRegressor(min_samples_split=3)
```

```python
min_samples_split_model.fit(X,y)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(10,7))
tree.plot_tree(min_samples_split_model,filled=True)
```

# ccp_alpha

```python
ccp_alpha_model=DecisionTreeRegressor(ccp_alpha=0.5) #heigher the value of ccp_aplha more pruning occurs.
```

```python
ccp_alpha_model.fit(X,y)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(10,7))
tree.plot_tree(ccp_alpha_model,filled=True)
```
